# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line_chart

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,color='r')
plt.xlabel('the data of week')
plt.ylabel('the price')
plt.title('Bit coin')
plt.show()
```

```python
plt.plot(x,y,color='r',linewidth=10)
plt.xlabel('the data of week')
plt.ylabel('the price')
plt.title('Bit coin')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'+',color='r')
plt.subplot(2,2,2)
plt.plot(x,y,'+',color='g')
```

# Legends

```python
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel('the data of week')
plt.ylabel('the price')
plt.title('Bit coin')
plt.legend(['squer','usd'])
plt.show()
```

# Barplot

```python
Courses=['Python','Java','C++','C']
values=[10,20,30,40]
plt.bar(Courses,values,color='maroon',alpha=.2)                  
plt.title('this is Bar')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.show()
```

```python
Courses=['Python','Java','C++','C']
values=[10,20,30,40]
plt.bar(Courses,values,color='maroon',alpha=.8)                  
plt.title('this is Bar')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.show()
```

# Bie_Chart

```python
cars=['Bmw','FORD','AUDI','TESLA','JAGURA','MERCEDES']
price=[23,17,35,29,21,41]
colors=['orange','cyan','brown','grey','indigo','beige']
plt.pie(price,labels=cars,colors=colors)
plt.show()
```

```python
cars=['Bmw','FORD','AUDI','TESLA','JAGURA','MERCEDES']
price=[23,17,35,29,21,41]
colors=['orange','cyan','brown','grey','indigo','beige']
plt.pie(price,colors=colors)
plt.legend(cars)
plt.show()
```

# Histogram

# LIneplot

```python
sns.lineplot(x=df['timepoint'],y=df['signal'])
```

```python
sns.lineplot(x=df['timepoint'],y=df['signal'],hue=df['region'])
#hue Like groupby
```

# Barplot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
sns.barplot(x=x,y=y)
```
